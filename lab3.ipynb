```python
collection.insert_one({
    "name": "Alice",
    "email": "alice@email.com",
    "country": "USA"
})
print("✅ Customer inserted successfully!")
```

```python
for customer in collection.find({"country": "USA"}):
    print(customer)
```

```python
collection.update_one(
    {"name": "Alice"},
    {"$set": {"email": "alice_new@email.com"}}
)
print("✅ Email updated successfully!")
```

```python
collection.delete_one({"name": "Alice"})
print("✅ Customer deleted successfully!")
```

```python
collection.insert_one({
    "name": "Bob",
    "email": "bob@email.com",
    "country": "UK",
    "address": {
        "city": "London",
        "zip": "SW1A"
    }
})
print("✅ Nested document inserted!")
```

```python
for customer in collection.find({"address.city": "London"}):
    print(customer)
```
